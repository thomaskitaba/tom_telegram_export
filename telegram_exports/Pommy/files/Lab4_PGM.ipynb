##**2. Structure learning in Bayesian networks**

###**2.1. Score-based Structure Learning**

**Scoring functions**

```python
import pandas as pd
import numpy as np
from pgmpy.estimators import BDeu, K2, BIC
from pgmpy.models import DiscreteBayesianNetwork

# create random data sample with 3 variables, where Z is dependent on X, Y:
data = pd.DataFrame(np.random.randint(0, 4, size=(5000, 2)), columns=list('XY'))
data['Z'] = data['X'] + data['Y']

bdeu = BDeu(data, equivalent_sample_size=5)
k2 = K2(data)
bic = BIC(data)

model1 = DiscreteBayesianNetwork([('X', 'Z'), ('Y', 'Z')])  # X -> Z <- Y
model2 = DiscreteBayesianNetwork([('X', 'Z'), ('X', 'Y')])  # Y <- X -> Z


print(bdeu.score(model1))
print(k2.score(model1))
print(bic.score(model1))

print(bdeu.score(model2))
print(k2.score(model2))
print(bic.score(model2))
```

```python
print(bdeu.local_score('Z', parents=[]))
print(bdeu.local_score('Z', parents=['X']))
print(bdeu.local_score('Z', parents=['X', 'Y']))
```

**Search strategies**

```python
from pgmpy.estimators import ExhaustiveSearch

es = ExhaustiveSearch(data, scoring_method=bic)
best_model = es.estimate()
print(best_model.edges())

print("\nAll DAGs by score:")
for score, dag in reversed(es.all_scores()):
    print(score, dag.edges())
```

```python
from pgmpy.estimators import HillClimbSearch

# create some data with dependencies
data = pd.DataFrame(np.random.randint(0, 3, size=(2500, 8)), columns=list('ABCDEFGH'))
data['A'] += data['B'] + data['C']
data['H'] = data['G'] - data['A']

hc = HillClimbSearch(data)
best_model = hc.estimate(scoring_method=BIC(data))
print(best_model.edges())
```

###**2.2. Constraint-based Structure Learning**

```python
from pgmpy.estimators import PC
from pgmpy.estimators.CITests import chi_square

data = pd.DataFrame(np.random.randint(0, 3, size=(2500, 8)), columns=list('ABCDEFGH'))
data['A'] += data['B'] + data['C']
data['H'] = data['G'] - data['A']
data['E'] *= data['F']

print(chi_square(X='B', Y='H', Z=[], data=data, significance_level=0.05))          # dependent
print(chi_square(X='B', Y='E', Z=[], data=data, significance_level=0.05))          # independent
print(chi_square(X='B', Y='H', Z=['A'], data=data, significance_level=0.05))       # independent
print(chi_square(X='A', Y='G', Z=[], data=data, significance_level=0.05))          # independent
print(chi_square(X='A', Y='G', Z=['H'], data=data, significance_level=0.05))       # dependent
```

```python
est  = PC(data)
pdag = est.estimate(significance_level=0.01)
print("PDAG edges:", pdag.edges())

dag  = pdag.to_dag()
print("DAG edges :", dag.edges())
```

##**2.3. Hybrid Structure Learning**

```python
from pgmpy.estimators import MmhcEstimator
from pgmpy.estimators import BDeu

data = pd.DataFrame(np.random.randint(0, 3, size=(2500, 8)), columns=list('ABCDEFGH'))
data['A'] += data['B'] + data['C']
data['H'] = data['G'] - data['A']
data['E'] *= data['F']

mmhc  = MmhcEstimator(data)

model = mmhc.estimate(
    scoring_method=BDeu(data, equivalent_sample_size=10),  # score-based phase
    significance_level=0.01,   # CI-test α for MMPC (skeleton)
    tabu_length=10,            # hill-climb meta-heuristic
)

print("MMHC DAG edges:", model.edges())
```

# **Exercise Solution: Structure Learning on Tips Dataset**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## **2. Loading and Preprocessing the Dataset**

```python
# Load and preprocess the tips dataset
tips_data = sns.load_dataset('tips')

# Discretize continuous variables
tips_data['total_bill_cat'] = pd.cut(tips_data['total_bill'], bins=3, labels=[0, 1, 2])
tips_data['tip_cat'] = pd.cut(tips_data['tip'], bins=3, labels=[0, 1, 2])
tips_data['size_cat'] = tips_data['size'].astype('category')

# Select and encode discrete variables
discrete_columns = ['sex', 'smoker', 'day', 'time', 'total_bill_cat', 'tip_cat', 'size_cat']
encoded_data = tips_data[discrete_columns].apply(lambda x: x.cat.codes)

print("Preprocessed data shape:", encoded_data.shape)
```

## **3. Learn structure using HillClimSearch with BIC and PC algorithm**

### Structure Learning with HillClimbSearch adn BIC

```python
# Learn structure using HillClimbSearch with BIC
hc_estimator = HillClimbSearch(encoded_data)
hc_model = hc_estimator.estimate(scoring_method=BIC(encoded_data))
print("HillClimbSearch + BIC edges:", hc_model.edges())
```

### Structure Learning with PC Algorithm

```python
# Learn structure using PC algorithm
pc_estimator = PC(encoded_data)
pc_model = pc_estimator.estimate(significance_level=0.05).to_dag()
print("PC Algorithm edges:", pc_model.edges())
```

## **4. Comparison of Both Learned Structures**

```python
# Compare the learned structures
hc_edges = set(hc_model.edges())
pc_edges = set(pc_model.edges())

common_edges = hc_edges.intersection(pc_edges)
hc_only = hc_edges - pc_edges
pc_only = pc_edges - hc_edges

print("\n=== COMPARISON RESULTS ===")
print(f"Common edges ({len(common_edges)}): {common_edges}")
print(f"Only in HillClimbSearch ({len(hc_only)}): {hc_only}")
print(f"Only in PC Algorithm ({len(pc_only)}): {pc_only}")
print(f"Jaccard Similarity: {len(common_edges)/len(hc_edges.union(pc_edges)):.3f}")
```

## **Conclusion**

In our analysis of the tips dataset, we found that both learning methods discovered meaningful patterns, but in different ways. HillClimbSearch with BIC built a more detailed network, finding multiple connections between variables as it searched for the best-scoring structure. The PC algorithm took a more cautious approach, only keeping relationships that passed strict statistical tests.

This shows us an important lesson: HillClimbSearch gives us a comprehensive picture but might include some weaker connections, while PC algorithm gives us a simpler, more reliable network. The choice depends on whether you want to explore all possible relationships or focus only on the strongest ones. Both methods agreed on the core patterns, confirming that things like bill amount really do influence tipping behavior.
